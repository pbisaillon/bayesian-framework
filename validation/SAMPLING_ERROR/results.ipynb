```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Data format
#Time (Average) Time (min) Time (max) Time (var)  
#Error (Average) Error(min) Error(max) Error (var) Particles Simulations
```

```python
enkf = np.loadtxt("enkf_results.dat")
pfenkf = np.loadtxt("pfenkf_results_noresampling.dat")
pf = np.loadtxt("pf_results.dat")
```

```python
plt.semilogy(enkf[:,8], enkf[:,0]*60.0, 'b.')
plt.semilogy(pf[:,8], pf[:,0]*60.0, 'r.')
plt.semilogy(pfenkf[:,8], pfenkf[:,0]*60.0, 'k.')
plt.legend(["enkf","pf","pfenkf"], loc="lower right")
plt.xlabel("Particles")
plt.ylabel("Average time (sec)")
plt.show()

plt.semilogy(enkf[:,8], np.sqrt(enkf[:,4]), 'b')
plt.semilogy(pf[:,8], np.sqrt(pf[:,4]), 'r')
plt.semilogy(pfenkf[:,8], np.sqrt(pfenkf[:,4]), 'k')
plt.legend(["enkf","pf","pfenkf"])
plt.xlabel("Particles")
plt.ylabel("Average Sqaure Error")
plt.show()

plt.semilogy(enkf[:,8], np.sqrt(enkf[:,7]), 'b')
plt.semilogy(pf[:,8], np.sqrt(pf[:,7]), 'r')
plt.semilogy(pfenkf[:,8], np.sqrt(pfenkf[:,7]), 'k')
plt.legend(["enkf","pf","pfenkf"])
plt.xlabel("Particles")
plt.ylabel("Var Square Error")
plt.show()

```
