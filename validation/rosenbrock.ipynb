```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

The rosenbrock pdf is defined by

$$
\pi(x,y) \propto \exp(-\frac{(1-x)^2 + 100(y-x^2)^2}{20})
$$

```python
#Analytical solution
#Area under the curve is 6.26761

x = np.arange(-8, 8, 0.1)
y = np.arange(-10, 30, 0.1)
xx, yy = np.meshgrid(x, y)
z = np.exp(-((1-xx)**2+100.0*(yy-xx**2)**2)/20.0)
h = plt.contour(x,y,z)
```

```python
#Load the MCMC chains
mh = np.loadtxt("rosenbrock_mh.dat")
ap = np.loadtxt("rosenbrock_ap.dat")
am = np.loadtxt("rosenbrock_am.dat")
dram = np.loadtxt("rosenbrock_dram.dat")
```

```python
#Plot the MCMC samples
thin = 100
plt.plot(mh[::thin,1], mh[::thin,2], 'k.')
plt.title("MH samples")
plt.xlabel("x")
plt.ylabel("y")
plt.contour(x,y,z)
plt.show()

plt.plot(ap[::thin,1], ap[::thin,2], 'k.')
plt.xlabel("x")
plt.ylabel("y")
plt.contour(x,y,z)
plt.title("MH samples after AP")
plt.show()

plt.plot(am[::thin,1], am[::thin,2], 'k.')
plt.xlabel("x")
plt.ylabel("y")
plt.contour(x,y,z)
plt.title("Adaptive metropolis")
plt.show()

plt.plot(dram[::thin,1], dram[::thin,2], 'k.')
plt.xlabel("x")
plt.ylabel("y")
plt.contour(x,y,z)
plt.title("DRAM")
plt.show()
```

```python
#Individual chains
thin = 10
plt.plot(mh[::thin,1], 'k.')
plt.plot(ap[::thin,1], 'r.')
plt.plot(am[::thin,1], 'b.')
plt.plot(dram[::thin,1], 'g.')
plt.show()
```

```python
#Plot the log likelihood (no prior) vs the sample number
thin = 500
plt.plot(mh[::thin,0], np.exp(mh[::thin,3]), 'k.')
plt.plot(ap[::thin,0], np.exp(ap[::thin,3]), 'r.')
plt.plot(am[::thin,0], np.exp(am[::thin,3]), 'b.')
plt.xlabel("Sample")
plt.ylabel("Likelihood")
plt.title("Likelihood vs sample")
plt.show()
```
